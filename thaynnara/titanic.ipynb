## 3. Feature engineering

- getting the passegenrs titles

```python
dataFrame_test_data = [dataFrame, test]

for dataset in dataFrame_test_data:
    dataset["Title"] = dataset.Name.str.extract(" ([A-Za-z]+)\.", expand=False)
```

```python
dataFrame.Title.value_counts()
```

```python
test.Title.value_counts()
```

```python
for dataset in dataFrame_test_data:
    title_map = {}
    for title in dataset["Title"]:
        
        if (title == "Mr"): 
            title_map["Mr"] = 0
        elif (title == "Miss"): 
            title_map["Miss"] = 1
        elif (title == "Mrs"): 
            title_map["Mrs"] = 2
        else: 
            title_map[title] = 3
    dataset.Title = dataset.Title.map(title_map)
        
```

```python
dataFrame.Title.value_counts()
```

```python
# absolute values of the people according with theirs titles
bar_chart("Title", False)
```

```python
def percent(valor):
    p = "%0.3f %%" %(valor * 100)
    return p

dataFrame[dataFrame.Survived == 1]["Title"].value_counts(normalize=True).apply(percent)
```

```python
# percent values of the people according with theirs titles
def percent2(valor):
    return (valor * 100)

def normalizeTitle(title):
    return (dataFrame[dataFrame.Title == title]["Survived"].value_counts(normalize=True).apply(percent2))

df = pandas.DataFrame({"Mr" : normalizeTitle(0),
                      "Miss" : normalizeTitle(1),
                      "Mrs" : normalizeTitle(2),
                      "others" : normalizeTitle(3)})

df.plot(kind="bar")
```

```python
print(normalizeTitle(0)) # Mr
print(normalizeTitle(1)) # Miss
print(normalizeTitle(2)) # Mrs
print(normalizeTitle(3)) # Others
```

### converting numerical age to categorical variable

- child: 0
- young: 1
- adult: 2
- mid-age: 3
- senior: 4

```python
for dataset in dataFrame_test_data:
    dataset.loc[dataset.Age <= 16, "Age"] = 0,
    dataset.loc[(dataset.Age > 16) & (dataset.Age <= 26), "Age"] = 1,
    dataset.loc[(dataset.Age > 26) & (dataset.Age <= 36), "Age"] = 2,
    dataset.loc[(dataset.Age > 36) & (dataset.Age <= 62), "Age"] = 3,
    dataset.loc[dataset.Age > 62, "Age"] = 4
```
