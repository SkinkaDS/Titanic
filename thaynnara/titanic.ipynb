#### Visualization

```python
import matplotlib.pyplot as plt
import seaborn as seaborn

%matplotlib inline
seaborn.set()
```

```python
# getting all survivors and the their sex and couting it
s = dataFrame[dataFrame.Survived == 1]["Sex"].value_counts()
s
```

```python
# getting the sex of the dead ones
d = dataFrame[dataFrame.Survived == 0]["Sex"].value_counts()
d
```

```python
df = pandas.DataFrame([s,d])
df
```

```python
df.index = ['Survived', 'Dead']
df
```

```python
df.plot(kind="bar", stacked=False, figsize=(8, 5), edgecolor="white", title= "Passagers")
```

```python
df.plot.bar(title= "Passagers",  stacked=True, figsize=(8, 5), edgecolor="white", color= ["red", "blue"], alpha=0.8 )
```

```python
def bar_chart(feature, s):
    
    survived = dataFrame[dataFrame.Survived == 1][feature].value_counts()
    dead = dataFrame[dataFrame.Survived == 0][feature].value_counts()
    df = pandas.DataFrame([survived, dead])
    df.index = ["Survived", "Dead"]
    df.plot(kind="bar", stacked=s, figsize=(10, 5))
```

```python
bar_chart("Sex", False)
```

```python
bar_chart("Age", True)
```

```python
bar_chart("Age", False)
```
